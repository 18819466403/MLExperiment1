```python
from numpy import *
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_svmlight_file
import matplotlib.pyplot as plt
import numpy
def getData():
    data = load_svmlight_file("C:/Users/YOUNA/Desktop/housing_scale.txt")
    return data[0], data[1]
def model(weigth,x):
    return x.dot(weight);

def loss(weight,x,y):  
    return ((model(weight,x)-y)**2).sum()/2
 
def gradiend(weight,x,y):    
    return x.T.dot(model(weight,x)-y)

x, y = getData()
x=x.toarray()
xTrain, xValidatuon, yTrain, yValiidation = train_test_split(x, y, test_size=0.5, random_state=42)
weight = numpy.zeros(xTrain.shape[1])

iteration = 100
learningRate = 0.001 
trainLoss = []
validateLoss = []
for i in range(iteration):
    grad = gradiend(weight,xTrain,yTrain)
    
    weight = weight - grad*learningRate
    trainLoss.append(loss(weight,xTrain,yTrain))
    validateLoss.append(loss(weight,xValidatuon,yValiidation))
 
plt.plot(range(iteration),trainLoss, label='train loss')
plt.plot(range(iteration),validateLoss, label='validtion loss')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.legend()
plt.show()
```
